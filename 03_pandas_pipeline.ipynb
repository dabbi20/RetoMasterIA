Durante la carga de la información se especifica la codificación UTF-8 para garantizar la correcta interpretación de caracteres especiales y se convierte automáticamente la columna **fecha** al tipo de dato fecha, facilitando posteriores análisis temporales.

Posteriormente, se utiliza el método **head()** para visualizar los primeros registros del conjunto de datos y verificar que la carga se haya realizado correctamente.

- Convertir la columna de fechas a un formato adecuado para análisis.

### Operaciones realizadas

La exploración inicial constituye una etapa esencial dentro del proceso de preparación de datos, ya que facilita la identificación de inconsistencias que podrían afectar los resultados del análisis posterior.

### Objetivo

### Métodos utilizados

| `info()` | Muestra información general del DataFrame, tipos de datos y cantidad de valores no nulos |

## Limpieza de datos y eliminación de valores nulos

Una vez identificados los registros con información incompleta, se procede a realizar el proceso de limpieza del conjunto de datos. Para ello se utiliza el método **dropna()**, eliminando aquellos registros que presentan valores nulos en campos considerados críticos para el análisis.

Las columnas seleccionadas corresponden a atributos esenciales del proceso de ventas, tales como identificadores, información del cliente, producto, cantidad, precio y vendedor. Mantener registros completos garantiza la consistencia de los cálculos y mejora la calidad de los resultados obtenidos.

### Objetivo

- Eliminar registros con información incompleta.
- Mejorar la calidad del conjunto de datos.
- Garantizar la consistencia de los análisis posteriores.
- Preparar los datos para las etapas de transformación y agregación.

### Columnas validadas

- order_id
- fecha
- cliente_id
- cliente_nombre
- region
- producto
- categoria
- cantidad
- precio_unitario
- descuento
- vendedor

### Procedimiento realizado

1. Identificación de columnas críticas.
2. Eliminación de registros con valores nulos.
3. Creación de una copia independiente del DataFrame limpio.
4. Comparación entre registros originales y registros finales.

````python
df_limpio = df.dropna(subset=[
    "order_id",
    "fecha",
    "cliente_id",
    "cliente_nombre",
    "region",
    "producto",
    "categoria",
    "cantidad",
    "precio_unitario",
    "descuento",
    "vendedor"
]).copy()

print("Registros originales:", len(df))
print("Registros eliminados:", len(df) - len(df_limpio))
print("Registros finales:", len(df_limpio))
````

## Conversión de tipos de datos y validación numérica

Después de eliminar los registros con valores nulos, se realiza una validación adicional sobre las columnas numéricas del conjunto de datos. Este proceso garantiza que los valores almacenados en las columnas **cantidad**, **precio_unitario** y **descuento** sean interpretados correctamente como datos numéricos.

La función **to_numeric()** permite convertir los valores al tipo de dato adecuado. El parámetro `errors="coerce"` transforma cualquier valor inválido o no numérico en valores nulos (NaN), facilitando su posterior identificación y eliminación.

Una vez realizada la conversión, se eliminan los registros que aún presenten valores faltantes en estas columnas críticas y se verifica nuevamente la estructura del DataFrame.

### Objetivo

- Garantizar la integridad de los datos numéricos.
- Detectar valores inconsistentes o mal formateados.
- Preparar el conjunto de datos para cálculos matemáticos.
- Validar la estructura final del DataFrame limpio.

### Columnas convertidas

| Columna | Tipo esperado |
|----------|--------------|
| cantidad | Numérico |
| precio_unitario | Numérico |
| descuento | Numérico |

### Procedimiento realizado

1. Conversión de columnas al formato numérico.
2. Transformación de valores inválidos en NaN.
3. Eliminación de registros con datos numéricos faltantes.
4. Verificación final mediante `info()`.

````python
df_limpio["cantidad"] = pd.to_numeric(df_limpio["cantidad"], errors="coerce")
df_limpio["precio_unitario"] = pd.to_numeric(df_limpio["precio_unitario"], errors="coerce")
df_limpio["descuento"] = pd.to_numeric(df_limpio["descuento"], errors="coerce")

df_limpio = df_limpio.dropna(subset=[
    "cantidad",
    "precio_unitario",
    "descuento"
]).copy()

df_limpio.info()
````

## Transformación de datos y creación de métricas de negocio

Una vez completado el proceso de limpieza y validación de los datos, se procede a generar nuevas variables que permitan realizar análisis comerciales más completos.

En esta etapa se calculan indicadores relacionados con los ingresos obtenidos por las ventas y se crea una columna temporal que facilitará la elaboración de reportes mensuales.

### Objetivo

- Calcular el ingreso bruto generado por cada venta.
- Calcular el ingreso neto considerando los descuentos aplicados.
- Crear una variable temporal para análisis por mes.
- Preparar los datos para procesos de agregación y generación de indicadores.

### Variables generadas

| Variable | Descripción |
|-----------|------------|
| ingreso_bruto | Resultado de multiplicar la cantidad vendida por el precio unitario |
| ingreso_neto | Ingreso bruto descontando el porcentaje de descuento aplicado |
| mes | Periodo mensual extraído de la fecha de la venta |

### Fórmulas utilizadas

**Ingreso Bruto**

```text
Ingreso Bruto = Cantidad × Precio Unitario
```

**Ingreso Neto**

```text
Ingreso Neto = Ingreso Bruto × (1 - Descuento)
```

### Procedimiento realizado

1. Cálculo del ingreso bruto por registro.
2. Aplicación de descuentos para obtener el ingreso neto.
3. Extracción del mes a partir de la fecha de venta.
4. Visualización de los primeros registros transformados.

````python
df_limpio["ingreso_bruto"] = (
    df_limpio["cantidad"] * df_limpio["precio_unitario"]
)

df_limpio["ingreso_neto"] = (
    df_limpio["ingreso_bruto"] * (1 - df_limpio["descuento"])
)

df_limpio["mes"] = df_limpio["fecha"].dt.to_period("M").astype(str)

df_limpio.head()
````

## Generación del resumen mensual de ventas

Con el propósito de obtener una visión consolidada del comportamiento comercial a lo largo del tiempo, se realiza una agrupación de los datos por mes utilizando la función **groupby()** de Pandas.

Posteriormente se aplican diferentes funciones de agregación para calcular indicadores clave de desempeño, incluyendo el número de pedidos, las unidades vendidas y los ingresos generados durante cada periodo.

Este tipo de análisis permite identificar tendencias temporales, evaluar el rendimiento de las ventas y facilitar la elaboración de reportes gerenciales.

### Objetivo

- Consolidar la información de ventas por mes.
- Calcular indicadores comerciales relevantes.
- Analizar la evolución de las ventas a lo largo del tiempo.
- Facilitar la generación de reportes ejecutivos.

### Indicadores calculados

| Indicador | Descripción |
|------------|------------|
| total_pedidos | Cantidad total de pedidos registrados en el mes |
| unidades_vendidas | Suma de unidades comercializadas |
| ingreso_bruto_total | Total de ingresos antes de descuentos |
| ingreso_neto_total | Total de ingresos después de aplicar descuentos |

### Funciones de agregación utilizadas

| Función | Propósito |
|----------|-----------|
| count() | Contar registros |
| sum() | Sumar valores numéricos |
| reset_index() | Restablecer el índice para generar un DataFrame estructurado |

````python
resumen_mensual = df_limpio.groupby("mes").agg(
    total_pedidos=("order_id", "count"),
    unidades_vendidas=("cantidad", "sum"),
    ingreso_bruto_total=("ingreso_bruto", "sum"),
    ingreso_neto_total=("ingreso_neto", "sum")
).reset_index()

resumen_mensual
````

## Evaluación del cumplimiento de metas por vendedor

Con el objetivo de analizar el desempeño comercial de cada vendedor, se construye una tabla de metas mensuales y posteriormente se compara con los ingresos netos generados durante el periodo analizado.

Para ello, primero se define un conjunto de metas comerciales asociadas a cada vendedor. Posteriormente, se calculan los ingresos netos acumulados mediante una agrupación de los datos y se combinan ambas fuentes de información utilizando la función **merge()**.

Finalmente, se calcula el porcentaje de cumplimiento de la meta asignada a cada vendedor, permitiendo evaluar el grado de logro de los objetivos comerciales establecidos.

### Objetivo

- Definir metas comerciales para cada vendedor.
- Calcular los ingresos netos generados por vendedor.
- Integrar información de ventas y objetivos comerciales.
- Determinar el porcentaje de cumplimiento de metas.

### Procedimiento realizado

1. Creación de la tabla de metas mensuales.
2. Agrupación de ventas por vendedor.
3. Cálculo de ingresos netos acumulados.
4. Integración de datos mediante `merge()`.
5. Cálculo del porcentaje de cumplimiento.

### Fórmula utilizada

**Porcentaje de Cumplimiento**

```text
( Ingreso Neto Total / Meta Mensual ) × 100
```

### Indicadores generados

| Indicador | Descripción |
|------------|------------|
| ingreso_neto_total | Total de ingresos netos obtenidos por el vendedor |
| meta_mensual | Objetivo comercial asignado |
| porcentaje_cumplimiento | Nivel de cumplimiento respecto a la meta |

````python
metas = pd.DataFrame({
    "vendedor": ["Ana López", "Carlos Ruiz", "Luis Mora", "María Torres"],
    "meta_mensual": [8000, 7500, 8000, 7000]
})

ventas_vendedor = df_limpio.groupby("vendedor").agg(
    ingreso_neto_total=("ingreso_neto", "sum")
).reset_index()

cumplimiento = pd.merge(
    ventas_vendedor,
    metas,
    on="vendedor",
    how="left"
)

cumplimiento["porcentaje_cumplimiento"] = (
    cumplimiento["ingreso_neto_total"] / cumplimiento["meta_mensual"]
) * 100

cumplimiento
````

## Carga de los datos procesados en SQLite

Una vez finalizadas las etapas de limpieza, transformación y análisis, se procede a almacenar los resultados en una base de datos SQLite. Este proceso corresponde a la fase **Load** dentro del flujo ETL (Extract, Transform, Load).

El almacenamiento de la información en una base de datos relacional permite conservar los datos procesados de forma estructurada y facilita la ejecución de consultas posteriores utilizando SQL.

Durante esta etapa se crean o actualizan tres tablas principales:

- **ventas:** contiene el conjunto de datos limpio y transformado.
- **resumen_mensual:** almacena los indicadores agregados por periodo mensual.
- **cumplimiento_vendedores:** registra los indicadores de desempeño y cumplimiento de metas por vendedor.

### Objetivo

- Persistir los datos procesados en una base de datos SQLite.
- Organizar la información en tablas especializadas.
- Facilitar consultas posteriores mediante SQL.
- Completar la fase de carga del proceso ETL.

### Tablas almacenadas

| Tabla | Descripción |
|---------|------------|
| ventas | Datos limpios y transformados |
| resumen_mensual | Indicadores agregados por mes |
| cumplimiento_vendedores | Indicadores de desempeño comercial |

### Procedimiento realizado

1. Creación de la conexión a SQLite.
2. Exportación del DataFrame principal.
3. Exportación del resumen mensual.
4. Exportación de indicadores de cumplimiento.
5. Confirmación de cambios mediante `commit()`.

````python
conn = sqlite3.connect("techventas.db")

df_limpio.to_sql(
    "ventas",
    conn,
    if_exists="replace",
    index=False
)

resumen_mensual.to_sql(
    "resumen_mensual",
    conn,
    if_exists="replace",
    index=False
)

cumplimiento.to_sql(
    "cumplimiento_vendedores",
    conn,
    if_exists="replace",
    index=False
)

conn.commit()
````
